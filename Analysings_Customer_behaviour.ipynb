### Which month has the highest sale/In which month customer spends the most

```python
data.head()
```

```python
data['month']=pd.to_datetime(data['DATE']).dt.month
```

```python
data.head()
```

```python
data['month']=data['month'].astype('str')
```

```python
look_up={'1':'Jan','2': 'Feb', '3': 'Mar', '4': 'Apr', '5': 'May',
           '6': 'Jun', '7': 'Jul', '8': 'Aug', '9': 'Sep' , '10': 'Oct' , '11': 'Nov', '12': 'Dec'}
data['month']=data['month'].apply(lambda x: look_up[x])
```

```python
data.head()
```

```python
month_data=data.groupby('month')['TOT_SALES'].sum()
```

```python
month_data=month_data.to_frame().reset_index()
```

```python
month_data
```

```python
sns.barplot(x='month',y='TOT_SALES',data=month_data)
```

### Which Product is sold the most in december? & which brand's products are sold the most?

```python
data.head()
```

```python
dec_data=data.loc[data['month'] == 'Dec']
```

```python
dec_data.head()
```

```python
pro_qty=dec_data.groupby('PROD_NAME')['PROD_QTY'].sum()
```

```python
pro_qty=pro_qty.to_frame().reset_index()
```

```python
pro_qty
```

```python
pro_qty.sort_values(by='PROD_QTY',ascending=False)
```

```python
brand_qty=dec_data.groupby('BRAND')['PROD_QTY'].sum()
```

```python
brand_qty=brand_qty.to_frame().reset_index()
```

```python
brand_qty.sort_values(by='PROD_QTY',ascending=False)
```

```python
jan_data=data.loc[data['month'] == 'Jan']
```
